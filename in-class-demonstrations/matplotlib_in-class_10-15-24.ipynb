```python
fig2, ax2 = plt.subplots(2, 3)
ax2[0, 0].scatter(x = [1, 2, 3, 4], y = [10, 11, 12, 13])
ax2[1, 1].scatter(x = [1, 2, 3, 4], y = [10, 11, 12, 13])
plt.show()
```

```python
fig2, ax2 = plt.subplots(4, 7)
ax2[0, 3].scatter(x = [1, 2, 3, 4], y = [10, 11, 12, 13])
ax2[0, 3].scatter(x = [1, 2, 3, 4], y = [10, 11, 12, 13], color = 'red',
                 alpha = .5)
ax2[1, 2].scatter(x = [1, 2, 3, 4], y = [10, 11, 12, 13])
ax2[3, 0].scatter(x = [1, 2, 3, 4], y = [10, 11, 12, 13])
ax2[3, 5].scatter(x = [1, 2, 3, 4], y = [10, 11, 12, 13])
plt.show()
```
