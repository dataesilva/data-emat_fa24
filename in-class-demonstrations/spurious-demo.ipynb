```python
import statsmodels.formula.api as smf
import pandas as pd
import seaborn as sns
```

```python
sal = pd.read_csv('Salary Data.csv')
sal.head()
```

```python
sal['ye'] = sal['Years of Experience']
```

```python
smf.ols('Salary ~ Age', data = sal).fit().summary()
```

```python
smf.ols('Salary ~ Age + ye', data = sal).fit().summary()
```
