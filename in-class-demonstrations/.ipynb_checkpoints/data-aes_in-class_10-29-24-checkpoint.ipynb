```python
fig2, ax2 = plt.subplots(2, 3)
ax2[0,0].hist(ad['danceability'], bins = 5)
ax2[0, 1].hist(ad['energy'], bins = 5)
ax2[0,2].hist(ad['acousticness'], bins = 5)
ax2[1,0].hist(ad['instrumentalness'], bins = 5)
ax2[1,1].hist(ad['valence'], bins = 5)
ax2[1,2].hist(ad['liveness'], bins = 5)

plt.show()
```

```python
phanto = ad[ad['artist-album'] == 'Phantogram: Memory of A Day']
stv = ad[ad['artist-album'] == 'St. Vincent: All Born Screaming']

fi4, ax4 = plt.subplots()
ax4.scatter(phanto['energy'], phanto['valence'], color = 'red')
ax4.scatter(stv['energy'], stv['valence'], color = 'gray')
plt.show()
```

```python
phanto = ad[ad['artist-album'] == 'Phantogram: Memory of A Day']
stv = ad[ad['artist-album'] == 'St. Vincent: All Born Screaming']

fi4, ax4 = plt.subplots()
ax4.scatter(phanto['danceability'], phanto['valence'], color = 'red')
ax4.scatter(stv['danceability'], stv['valence'], color = 'gray')
plt.show()
```

```python
phanto = ad[ad['artist-album'] == 'Phantogram: Memory of A Day']
stv = ad[ad['artist-album'] == 'St. Vincent: All Born Screaming']


fig4, ax4 = plt.subplots()
phanto_scatter = ax4.scatter(phanto['danceability'], phanto['valence'], color = 'red', label = 'a')
stv_scatter = ax4.scatter(stv['danceability'], stv['valence'], color = 'gray', label = 'b') #, label = "St. Vincent")
ax4.set_xlabel('Danceability')
ax4.set_ylabel('Valence')
ax4.set_title('Phantogram and St. Vincenet: Danceability by Valence')
ax4.legend(['Phantogram = red', 'St. Vincent = grey'])
fig4.suptitle('Something')

plt.style.use('grayscale')

plt.legend(handles = [phanto_scatter, stv_scatter])
plt.show()


# plt.xlabel('Something')
# plt.title('Something')
```
