```python
import pandas as pd
from matplotlib import pyplot as plt
```

```python
ad = pd.read_csv('../audio-features-analysis.csv')
ad.head()
```

```python
ad.groupby('artist-album')['id'].count()
```

```python
fig1, ax1 = plt.subplots()
ax1.hist(ad['danceability'], bins = 6)

plt.show()
```

```python
fig2, ax2 = plt.subplots(2, 3)
ax2[0,0].hist(ad['danceability'], bins = 5)
ax2[0, 1].hist(ad['energy'], bins = 5)
ax2[0,2].hist(ad['acousticness'], bins = 5)
ax2[1,0].hist(ad['instrumentalness'], bins = 5)
ax2[1,1].hist(ad['valence'], bins = 5)
ax2[1,2].hist(ad['liveness'], bins = 5)

plt.show()
```

```python
ad[ad['liveness'] > .35][['name', 'artist-album']]
```

```python
fig3, ax3 = plt.subplots(1, 3, figsize = (8, 2))
ax3[0].scatter(ad['danceability'], ad['energy'])
ax3[1].scatter(ad['danceability'], ad['valence'])
ax3[2].scatter(ad['energy'], ad['valence'])
# ax3[2].axline((0, .2), .1)

plt.show()
```

```python
phanto = ad[ad['artist-album'] == 'Phantogram: Memory of A Day']
stv = ad[ad['artist-album'] == 'St. Vincent: All Born Screaming']

fi4, ax4 = plt.subplots()
ax4.scatter(phanto['energy'], phanto['valence'], color = 'red')
ax4.scatter(stv['energy'], stv['valence'], color = 'gray')
plt.show()
```
