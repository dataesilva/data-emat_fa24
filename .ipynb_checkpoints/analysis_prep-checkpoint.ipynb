```python
import urllib
import requests
import pandas as pd
import json
import base64
```

```python
def get_session_token(SessionID, SessionKey):
    url = 'https://accounts.spotify.com/api/token'
    data = {'grant_type':'client_credentials'}
    encoded_key = base64.b64encode(str(SessionID + ":" + SessionKey).\
                                   encode("ascii"))
    header = {'Authorization': 'Basic {}'.format(encoded_key.decode("ascii"))}
    response = requests.post(url,
                            data = data,
                            headers = header)
    print(response.status_code)
    return response.json()['access_token']
```

```python
t_features_ep = 'https://api.spotify.com/v1/audio-features'
tracks_ep = 'https://api.spotify.com/v1/tracks'
ab_tracks_ep = 'https://api.spotify.com/v1/albums/{}/tracks'
```

```python
def api_call(endpoint_url, api_header):
    response = requests.get(endpoint_url, headers = api_header)
    print(response.status_code)
    return response.json()
```

```python
session_header = {'Authorization': 'Bearer {}'.format(access_token)}
```

```python
stv_album_id = '3nRlJXz5W39luXRto5hc4f'
phanto_album_id = '4H1TE5m2eltLhVKVuB53b1'
```

```python
def album_tracks_audiofeatures(album_id, artist_album):
    album_response = api_call(ab_tracks_ep.format(album_id), session_header)
    album_df = pd.DataFrame(album_response['items'])
    track_ids = ','.join(album_df['id'].to_list())
    track_features = api_call(t_features_ep + '?ids={}'.format(track_ids),
                             session_header)
    track_info = api_call(tracks_ep + '?market=US&ids={}'.format(track_ids),
                             session_header)
    features_df = pd.DataFrame(track_features['audio_features'])
    tracks_df = pd.DataFrame(track_info['tracks'])
    merged = pd.merge(features_df, tracks_df,
                     how = 'inner', on = 'id')
    merged['artist-album'] = str(artist_album)
    return merged
```

```python
stv_merged = album_tracks_audiofeatures(stv_album_id, "St. Vincent: All Born Screaming")
stv_merged.head()
```

```python
phanto_merged = album_tracks_audiofeatures(phanto_album_id, "Phantogram: Memory of A Day")
phanto_merged.head()
```

```python
artist_compare = pd.concat([stv_merged, phanto_merged], axis = 0)
artist_compare
```
