```python
import pandas as pd
from matplotlib import pyplot as plt
```

```python
dat = pd.read_csv('cces_example_data.csv')
```

```python
dat.columns
```

```python
dat['pid3'].value_counts()
```

```python
fig1, ax1 = plt.subplots()
ax1.bar(dat['pid3'].value_counts().index, dat['pid3'].value_counts())
plt.show()
```

```python
fig2, ax2 = plt.subplots()
ax2.bar(['Democratic Party', 'Republican Party', 'Independent'], dat['pid3'].value_counts())
plt.show()
```

```python
fig3, ax3 = plt.subplots()
ax3.bar(['Democratic Party', 'Republican Party', 'Independent'], dat['pid3'].value_counts().sort_values())
plt.show()
```

```python
party_pref = dat['pid3'].value_counts().sort_values(ascending = False)
fig4, ax4 = plt.subplots()
ax4.bar(party_pref.index, party_pref)
plt.show()
```
